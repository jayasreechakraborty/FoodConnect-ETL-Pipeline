```python
#feature engineering#
```

```python
#Food Consumed
```

```python
df_clean['food_consumed'] = df_clean['quantity_of_food'] - df_clean['food_wastage_amount']
```

```python
df_clean.head()
```

```python
#Waste Percentage
```

```python
df_clean['waste_percentage'] = (
    df_clean['food_wastage_amount'] /
    df_clean['quantity_of_food']
) * 100
```

```python
df_clean['waste_percentage'] = df_clean['waste_percentage'].round(2)
```

```python
#Price Mapping
```

```python
price_map = {
    'Low': 120,
    'Moderate': 220,
    'High': 350
}
```

```python
#Revenue Loss
```

```python
df_clean['price_per_kg'] = df_clean['pricing'].map(price_map)
```

```python
df_clean['revenue_loss'] = (
    df_clean['food_wastage_amount'] *
    df_clean['price_per_kg']
)
```

```python
#Donation Possible
```

```python
df_clean['donation_possible'] = np.where(
    df_clean['food_wastage_amount'] <= 30,
    'Yes',
    'No'
)
```

```python
#Waste Category
```

```python
def waste_level(x):
    if x < 20:
        return "Low"
    elif x < 35:
        return "Medium"
    else:
        return "High"
```

```python
df_clean['waste_category'] = df_clean['food_wastage_amount'].apply(waste_level)
```

```python
#Event Size
```

```python
def event_size(x):
    if x < 250:
        return "Small"
    elif x < 400:
        return "Medium"
    else:
        return "Large"
```

```python
df_clean['event_size'] = df_clean['number_of_guests'].apply(event_size)
```

```python
#Save the Final Dataset
```

```python
df_clean.to_csv("food_waste_final.csv", index=False)
```
